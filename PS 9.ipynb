```python
import pandas as pd
import numpy as np

df = pd.read_csv('dataset_Facebook.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
mean = df['Paid'].mean()

median = df['Paid'].median()

mode = df['Paid'].mode()

print("The Mean is "+  str(mean) +".\nThe Median is "+ str(median) +".\nThe Mode is "+ str(mode) +".")
```

```python
df1 = df.fillna(value="mode")
df1.isnull().sum()
```

```python
df1.sort_values(by="Type", inplace=True)

df1.head(200)
```

```python
df_link = df1.loc[df1['Type']=="Link"]
df_link
```

```python
df_status = df1.loc[df1["Type"]=="Status"]
df_status
```

```python
df_photos = df1.loc[df1['Type']=='Photo']
df_photos
```

```python
df_link_status = pd.concat([df_link, df_status], axis=0)
df_link_status
```

```python
df_status_photo = pd.concat([df_status, df_photos], axis=0)
df_status_photo
```

```python
df_sorted = df1.sort_values(by='Page total likes')
df_sorted
```

```python
df_transpose = df_sorted.T
df_transpose
```

```python
df_melted = pd.melt(df_sorted, id_vars=['Type'], value_vars=['Page total likes'], value_name="Likes per Type")
df_melted
```

```python
df_melt1 = pd.melt(df_sorted, id_vars=['Type'], value_vars=['Post Hour', 'Page total likes'], value_name="values")
df_melt1
```

```python
df_casted = df_melt1.pivot_table(index="Type", columns='variable', values='values')
df_casted
```

```python
df_cast = pd.pivot_table(df_melt1, values='values', index='Type', columns='variable', aggfunc='first')
df_cast
```
